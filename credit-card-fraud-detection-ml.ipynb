# **Importing necessary Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import gridspec
```

# **Loading the Data**

```python
data = pd.read_csv("/kaggle/input/creditcardfraud/creditcard.csv")
print(data.head())
```

```python
print(data.describe())
```

# **Analyzing Class Distribution**

```python
fraud = data[data['Class'] == 1]
valid = data[data['Class'] == 0]

outlierFraction = len(fraud)/float(len(valid))

print(outlierFraction)
print('Fraud Cases : {}'.format(len(fraud)))
print('Valid Transactions : {}'.format(len(valid)))
```

# **Exploring Transaction Amounts**

```python
print("Amount details of the fraudulent transaction")
fraud.Amount.describe()
```

```python
print("Amount details of the valid transaction")
valid.Amount.describe()
```

# **Plotting Correlation Matrix**

```python
corrmat = data.corr()
fig = plt.figure(figsize = (12,12))
sns.heatmap(corrmat, vmax = .8, square = True)
plt.show()
```

```python
X = data.drop(['Class'], axis = 1)
Y = data["Class"]
print(X.shape)
print(Y.shape)

xData = X.values
yData = Y.values

from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest = train_test_split(
    xData, yData, test_size = 0.2, random_state = 42
)
```

# **Building and Training the Model**

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier()
rfc.fit(xTrain, yTrain)

yPred = rfc.predict(xTest)
```

```python
import joblib
import time
from sklearn.ensemble import RandomForestClassifier

# Mesurer le temps d'entraînement
start_train = time.time()

rfc = RandomForestClassifier(
    n_estimators=500,          # Number of trees in the forest (more trees = more stable, but slower to train)
    max_depth=20,              # Maximum depth of each tree (limits tree complexity to avoid overfitting)
    min_samples_split=2,       # Minimum number of samples required to split an internal node (2 = very flexible)
    min_samples_leaf=1,        # Minimum number of samples required to be at a leaf node (1 = very detailed leaves, may overfit)
    max_features='sqrt',       # Number of features to consider when looking for the best split ('sqrt' is common for classification)
    bootstrap=True,            # Whether bootstrap samples are used when building trees (True = use sampling with replacement)
    random_state=42,           # Random seed for reproducibility
    n_jobs=-1                  # Use all available CPU cores for training (parallel processing)
)

rfc.fit(xTrain, yTrain)

end_train = time.time()
print(f"Temps d'entraînement : {end_train - start_train:.2f} secondes")

# Mesurer le temps de prédiction
start_pred = time.time()

yPred = rfc.predict(xTest)

end_pred = time.time()
print(f"Temps de prédiction : {end_pred - start_pred:.2f} secondes")

# Save the trained model to a file
joblib.dump(rfc, 'random_forest_model_fraud_classification.pkl')
```

# **Evaluating the Model**

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, matthews_corrcoef, confusion_matrix 

accuracy = accuracy_score(yTest, yPred)
precision = precision_score(yTest, yPred)
recall = recall_score(yTest, yPred)
f1 = f1_score(yTest, yPred)
mcc = matthews_corrcoef(yTest, yPred)

print("Métriques d’évaluation du modèle :")
print(f"Accuracy : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall : {recall:.4f}")
print(f"F1_score : {f1:.4f}")
print(f"Matthews Correlation Coefficient: {mcc:.4f}")

conf_matrix = confusion_matrix(yTest, yPred)
plt.figure(figsize=(8, 8))

sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues",
           xticklabels=['Normal', 'Fraud'], yticklabels=['Normal', 'Fraud'])

plt.title("Confusion Matrix")
plt.xlabel("Prediction")
plt.ylabel("True Class")
plt.show()
```
